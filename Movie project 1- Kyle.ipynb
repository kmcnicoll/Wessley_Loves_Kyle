```python
tvs_url=[]

for a in list(range(1,501)):
    page = requests.get("https://www.themoviedb.org/tv"+"?page="+str(a))
    soup = BeautifulSoup(page.content, 'html.parser')
    for elm in  soup.find_all(class_="view_more"):
        for a_elm in elm.find_all("a"):
            tvs_url.append(a_elm.attrs["href"])

```

```python
len(tvs_url)
#URLS FOR 10,000 tv shows
```
